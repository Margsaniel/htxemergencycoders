```python
# Dependencies and setup
%matplotlib inline
import os
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import csv
reservoirs_and_bayou_df = pd.read_csv('../data/clean_data/reservoirs_and_bayou.csv')
reservoirs_and_bayou_df.head()
```

```python
# Set index of dataframe to be "Date" for year 2017
reservoirs_and_bayou_df.set_index("Date").head()
```

```python
# Create a list of dates and elevations for Addicks Reservoir
x = reservoirs_and_bayou_df["Date"]
y1 = reservoirs_and_bayou_df["Addicks elevation (ft)"]
y2 = reservoirs_and_bayou_df["Addicks rain (in)"]

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.plot(x, y1, "b-")
ax2.plot(x, y2, "r-")

ax1.set_xlabel("")
ax1.set_ylabel("ELEVATION  (FT)", color="b")
ax2.set_ylabel("RAIN  FALL  (IN)", color="r", rotation=270, labelpad=15)

plt.title("2017 Addicks Reservoir Elevation and Rain Fall")

frequency = 30

plt.xticks(x[::frequency],rotation=45)
plt.grid()

#plt.axhline(y=95, color='r', linestyle='-')
#plt.text(95, 95, "Top of Spillway", fontsize=10, va='center', ha='center', backgroundcolor='w')
#plt.axhline(y=88, color='r', linestyle='-')
#plt.text(88, 88, "Highway 6 Flooding", fontsize=10, va='center', ha='center', backgroundcolor='w')

plt.savefig("../images/addicks_reservoir_elevation.png")
plt.show()
```
